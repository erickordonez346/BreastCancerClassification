```python
import pandas as pd
import matplotlib as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('breastaware.csv')
df.head(10)
```

```python
# if malignant, we will give it a value of 1, else a value of 0

df['new_diagnosis'] = df['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)
df.head()
```

```python
# we drop any rows that have at least one missing value
df.dropna()
df.describe()
```

```python
# we need to drop columns that don't mean anything
df.drop(['Unnamed: 32'], axis = 1, inplace = True)
```

```python
# let's create a function that will normalize and update all the columns that
# need normalizing
# we do this so no one value can overinfluence the distance when the
# knn model computes it
column_list = [x for x in df.columns]
columns_that_need_normalizing = column_list[2:-3]

def normalize_all_columns(list_columns, dataframe):
    scalar = MinMaxScaler()
    for column in list_columns:
        new_column_data = [[x] for x in dataframe[column]]
        scalar.fit(new_column_data)
        df['{}'.format(column)] = scalar.transform(new_column_data)
normalize_all_columns(columns_that_need_normalizing, df)
df.describe()
```

```python
# now we need to to put all of the values for each id into a list so that
# we can train the KNN model
# one huge list of all the data
data_for_all_ids = []
for patient in range(0,569):
    id_data = list(df.iloc[patient][2:])
    data_for_all_ids.append(id_data)
labels_for_all_ids = [x for x in df['new_diagnosis']]
```

```python
# now let's call the knn model

classifier = KNeighborsClassifier(n_neighbors = 5)

# we will use 80 percent of the data for training and 20 percent for testing
training_data, test_data, training_labels, test_labels = train_test_split(data_for_all_ids, labels_for_all_ids, test_size = 0.2,random_state = 100)
classifier.fit(training_data, training_labels)
print('This is the accuracy of the model:')
print(classifier.score(test_data, test_labels))
```
